```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
y = x**2
```

# PyPlot
PyPlot est un module pour tracer des graph. Ne pas confondre avec une approche Orientée Objet
## Instructions de base

```python
plt.figure(figsize=(8, 5))    # Initialize la zone de graph 
                               # Option figsize (facultatif) = dimensions en inch
plt.title('Titre')
plt.xlabel('Axe X')
plt.ylabel('Axe Y')
plt.plot(x, y)    # Courbe (a partir des points ayant pour coord contenues dans x & y)
plt.scatter(x, y, marker='|')   # Nuage de points
plt.fill_between(x, x-1, x+1, alpha=0.2)   # Colorier une zone

# Options communes
plt.plot(x, x**3, label='Curv1', lw=3, ls='--', c='red', alpha=0.5)
# label: nom de la courbe
# lw: line width
# ls: line style (--, .., :)
# c: color (g, r, b, red, green, blue, black, etc.)
# alpha: transparence

# Option pour scatter
# s: size (taille des points)
# marker: forme des points (o, +, |, x)

plt.legend()   # Affiche la légende (doit être placé après les courbes)
plt.show()        # Affiche le graph
plt.savefig('figure.png')   # sauvegarde la courbe en tant qu'image
```

## Subplot

```python
plt.figure()
plt.subplot(2, 1, 1)    # Divise la zone de graph en 2 lignes, 1 colonne et active
                        # la 1ère zone pour les instructions suivantes
plt.title('y=x')
plt.plot(x, x)
plt.subplot(2, 1, 2)    # Divise la zone de graph en 2 lignes, 1 colonne et active
                        # la 2e zone pour les instructions suivantes
plt.title('y=x^2')
plt.plot(x, x**2)
```

# Approche OO

```python
fig, ax = plt.subplots()    # Renvoie 2 objets sur lesquels on va appeler nos méthodes. Note le S à subplotS
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)   # Si on fait des subplot, on peut partager 1 des 2 axes
ax[0].plot(x, x**2)            # Attention, si plusieurs subplots, ax est un NDarray
ax[1].plot(x, x**3)
plt.show()
```

# Graphiques avancés
## Nuages de points

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data
y = iris.target
names = iris.target_names
```

```python
plt.scatter(x[:,0], x[:,1], c=y, alpha=0.5, s=x[:,2]*10)
# Pour la couleur et la taille, on n'utilise pas une valeur constant mais une variable
# Tous les points apartenant à une même catégorie (target) ont la même couleur
# La taille du point est fonction d'un autre paramètre des données


plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
```

## Histogrammes

```python
plt.hist(x[:,0], bins=30)
plt.hist(x[:,1], bins=30)
# bins: nombre de barre de l'histogramme
plt.show()
```

```python
plt.hist2d(x[:,0], x[:,1], cmap='Blues')
plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
plt.colorbar()
plt.show()
```

## Lignes de niveau

```python
f = lambda x, y: np.sin(x) + np.cos(x+y) * np.cos(x)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
plt.contour(X, Y, Z, 20, colors='black')
plt.show()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.show()
```

# Graphiques 3D

```python
%matplotlib
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c=y)
plt.show()
```

```python
f = lambda x, y: np.sin(x) + np.cos(x+y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```
