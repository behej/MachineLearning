## Graphique de catégories
Des graphiques utiles lorsque nos données contiennent des catégories

````python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'embark_town', 'who', 'adult_male', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
````

````python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
````

````python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
````

# Graphiques de distribution

````python
sns.distplot(titanic['fare'])
````

````python
sns.jointplot(x='age', y='fare', data=titanic, kind='kde')
# kind: kde, hex
````

````python
sns.heatmap(titanic.corr())
````
